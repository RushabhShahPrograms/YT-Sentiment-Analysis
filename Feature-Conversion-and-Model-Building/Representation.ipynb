```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv('New_Merge_Data_Comment.csv')

# Count the number of comments for each class
class_counts = df['Class'].value_counts()

# Create a bar graph
class_counts.plot(kind='bar', figsize=(10, 6), rot=0)
plt.xlabel('Class', fontsize=14)
plt.ylabel('Number of Comments', fontsize=14)
plt.title('Distribution of Comments by Class', fontsize=16)
plt.tight_layout()
plt.show()

csv_file = 'New_Merge_Data_Comment.csv'
df.to_csv(csv_file, index=False)
```

```python
from wordcloud import WordCloud
text = df['Comment'].values 

wordcloud = WordCloud(background_color='white',collocations=False).generate(str(text))

plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```
